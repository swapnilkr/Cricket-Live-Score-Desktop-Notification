```python
# import library
import requests
from bs4 import BeautifulSoup
from win10toast import ToastNotifier
import re
from time import sleep
from __future__ import print_function

def start():
    print("LIVE CRICKET MATCHES LIST")
    
# Function for Windows toast desktop notification
def notification(score):
    notify = ToastNotifier()
    notify.show_toast(score,"Live Score",duration=10)
    
def sendmessage(message):
    subprocess.Popen(['notify-send',message])
    return

def pingurl(url):
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data)
    return soup

# returns only the live matches and ignores the completed ones
def currentMatches():
    # Function to fetch the live matches currently in play
    feed_page = requests.get('http://static.cricinfo.com/rss/livescores.xml') 
    xml_tree_soup = BeautifulSoup(feed_page.text,'lxml') 
    matches = xml_tree_soup.find_all('description') 
    live_matches = [similar.get_text() for similar in matches if '*' in similar.get_text()] 
    return live_matches

# Function to return the live score of the match specified
def getScore(numMatch):
    
    feed_page = requests.get('http://static.cricinfo.com/rss/livescores.xml')
    xml_tree_soup = BeautifulSoup(feed_page.text,'lxml')
    matches = xml_tree_soup.find_all('description')
    live_matches = [similar.get_text() for similar in matches if '*' in similar.get_text()]
    return live_matches[numMatch]
  

if __name__ == "__main__":
    start()
    match = currentMatches()
    print('Current matches in play')
    print('*'*50)
    # printing a list of all the current live matches 
    for i,match in enumerate(match):
                print('[{}] '.format(i) + 
                        re.search('\D+',match.split('v')[0]).group() + 'vs.' +
                        re.search('\D+',match.split('v')[1]).group()
                      )
    print('*'*50)
    print()
    numMatch = int(input('Pick the match number [0,1,2...] => '))


    # show desktop notification every 30 seconds
    while True:
        currScore = getScore(numMatch)
        notification(currScore)
        sleep(30)
```
